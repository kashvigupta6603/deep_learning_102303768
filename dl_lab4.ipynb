```python
n_features = X.shape[1]

w = np.zeros(n_features)
b = 0.0

print("Initial weights:", w)
print("Initial bias:", b)

```

```python
def predict(X, w, b):
    return X.dot(w) + b



```

Why is there no activation function?

What kind of values can
y_hat
 take?

How is this different from logistic regression?

ANSWER:

1.No activation function is needed because income is continuous.

2.any real number

3.Linear regression predicts any real number (−∞ to +∞), while logistic regression predicts a probability between 0 and 1 using a sigmoid function.

```python
def mean_squared_error(y, y_hat):
    return ((y_hat - y) ** 2).mean()



```

Why square the error?

What happens if one prediction is very wrong?

Why not just take absolute error

ANSWER:

1.Removes sign and
Punishes big mistakes heavily

2.That one example will dominate loss.
Model will try hard to fix it.

3.We don’t use absolute error because it’s not smooth at zero, making gradient descent harder to optimize

```python
def compute_gradients(X, y, y_hat):
    N = len(y)

    dw = (2/N) * X.T.dot(y_hat - y)
    db = (2/N) * (y_hat - y).sum()

    return dw, db

```

Why does X appear in dw but not in
db
?

Why does the error term appear everywhere?

What happens if error is zero?

ANSWER:

1. Weight controls how strongly a feature affects output.Bias is independent of feature

2.error term appears everywhere since gradients are directly driven by how wrong the prediction is

3. Training stops automatically.
Model has learned perfectly.

```python
def update_parameters(w, b, dw, db, lr):
    w = w - lr * dw
    b = b - lr * db
    return w, b
```

```python
lr = 0.01   # I first tried a larger learning rate
epochs = 1000

loss_history = []

for epoch in range(epochs):

    y_hat = predict(X, w, b)

    loss = mean_squared_error(y, y_hat)
    loss_history.append(loss)

    dw, db = compute_gradients(X, y, y_hat)

    w, b = update_parameters(w, b, dw, db, lr)

    if epoch % 100 == 0:
        print(f"Epoch {epoch}, Loss: {loss}")

```

Does loss decrease over time?

What happens if it increases?

How do learning rate and epochs interact?

ANSWER:

1.gradually decrease

2.if it increase learning rate is too high.

3.Learning rate controls how big each update step is, while epochs control how many total steps you take


```python
# resetting parameters
w = np.zeros(n_features)
b = 0.0

lr = 0.0001   # reduced learning rate

loss_history = []

for epoch in range(epochs):

    y_hat = predict(X, w, b)

    loss = mean_squared_error(y, y_hat)
    loss_history.append(loss)

    dw, db = compute_gradients(X, y, y_hat)

    w, b = update_parameters(w, b, dw, db, lr)

    if epoch % 100 == 0:
        print(f"Epoch {epoch}, Loss: {loss}")

```

```python
print("Final weights:", w)
print("Final bias:", b)

new_candidate = np.array([68,4.5])
predicted_salary = new_candidate.dot(w) + b

print("Predicted salary:", predicted_salary)

```
